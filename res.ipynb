```python
import pandas as pd
```

```python
df = pd.read_csv('res.csv')
```

```python
df.head()
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df['duration_seconds'] = (df['end_time'] - df['start_time']).dt.total_seconds()

df.sort_values(by='duration_seconds', ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['threads'],df['duration_seconds'], marker='o', linestyle='-', color='b')
plt.ylabel('Время, (сек)')
plt.xlabel('Количество воркеров')
plt.title('Время выполнения в зависимости от количества воркеров')
plt.grid(True)
plt.show()
```
